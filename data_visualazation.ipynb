## **Introduction to Data Visualization**

`Data visualization` is the process of presenting information and data in a graphical format. Utilizing visual elements such as charts, graphs, and maps, data visualization tools offer an intuitive approach to interpreting trends, outliers, and patterns within data.

```python
# import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "a":[1,2,3,4,5],
    "b": [1,4,9,16,25]
})

#lot datapoint fpr (x,y)
plt.plot(df["a"], df["b"])

#set the title of the plot
plt.title("Title goes here")

# label the x-axis
plt.xlabel('Label for A')

#label the y-axis
plt.ylabel('Label for B')


#Display the plot
plt.show()
```

### Types of Charts
- Bar
- Pie
- Line
- Scatter

```python
# generating bar chart
plt.bar(df['a'], df['b']);

# Set the title of the plot
plt.title('My first bar chart')
# Label the x-axis
plt.xlabel('A')
# Label the y-axis
plt.ylabel('B')

plt.show()
```

```python
# generating a pie chart
plt.pie(df['a'], labels=df['a'])

#set the title
plt.title('Title goes here')

plt.show()
```

```python
combined = pd.concat([df["a"], df["b"]])
# Generating a pie chart
plt.pie(combined, labels=combined, colors=['tomato', 'cornflowerblue', '#77BFE2', 'blue', 'crimson'])
# Set the title of the plot
plt.title('My first pie chart')
plt.show()
```

```python
plt.scatter(df["a"], df["b"])
plt.show()
```

```python
# Create subplot using Matplotlib (2x2)
fig, ax = plt.subplots(2, 2, figsize=(16, 12))

ax[0, 0] = plt.scatter(df["a"], df["b"])
ax[0, 1] = plt.scatter(df["a"], df["b"])
ax[1, 0] = plt.scatter(df["a"], df["b"])
ax[1, 1] = plt.scatter(df["a"], df["b"])

plt.tight_layout()

plt.show()
```

```python
combined = pd.concat([df["a"], df["b"]])

# Generating a pie chart
plt.pie(combined, labels=combined, colors=['tomato', 'cornflowerblue', '#77BFE2', 'blue', 'crimson'], autopct="%1.1f%%")

# Set the title of the plot
plt.title('My first pie chart')
plt.show()
```
